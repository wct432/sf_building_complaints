```python
def tokenize(sentences, tokenizer):
    input_ids, input_masks = [],[]
    for sentence in tqdm(sentences):
        inputs = tokenizer.encode_plus(sentence, add_special_tokens=True, 
                                       max_length=128,truncation=True,padding='max_length',
                                       return_attention_mask=True,return_token_type_ids=True)
        input_ids.append(inputs['input_ids'])
        input_masks.append(inputs['attention_mask'])      

    return (tf.convert_to_tensor(input_ids), tf.convert_to_tensor(input_masks))
```

```python
distilbert_tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

X_train_input_ids, X_train_input_masks = tokenize(X_train,distilbert_tokenizer)
X_val_input_ids, X_val_input_masks = tokenize(X_val,distilbert_tokenizer)
X_test_input_ids, X_test_input_masks = tokenize(X_test,distilbert_tokenizer)
```

```python
DISTILBERT_DROPOUT = 0.2
DISTILBERT_ATT_DROPOUT = 0.2
 
# Configure DistilBERT's initialization
config = DistilBertConfig(dropout=DISTILBERT_DROPOUT, 
                          attention_dropout=DISTILBERT_ATT_DROPOUT, 
                          output_hidden_states=True)
                          
#pre-trained DistilBERT model outputting raw hidden-states 
#needs head for classification
distilbert = TFDistilBertModel.from_pretrained('distilbert-base-uncased', config=config)

# Make DistilBERT layers untrainable
for layer in distilbert.layers:
    layer.trainable = False
```

```python
MAX_LENGTH = 128
LAYER_DROPOUT = 0.2
LEARNING_RATE = 1e-3
RANDOM_STATE = 42

def build_model(transformer, num_classes, max_length=MAX_LENGTH):
    
    """""""""
    Builds a BERT model for classification tasks using a Hugging Face 
    transformer with no head attached.
    
    Input:
      - transformer:  base Hugging Face transformer with no head.
      - max_length:   Controls the maximum number of encoded tokens in 
                      a sequence.
    
    Output:
      - model:        a compiled tf.keras.Model with added multi-class 
                      classification layerson top of the base Hugging Face 
                      transformer. 
    """""""""""
    
    #define metrics to monitor
    metrics = [
                tf.keras.metrics.CategoricalAccuracy(name='accuracy'),
                tf.keras.metrics.AUC(name='auc'),
    ]

    # define weight initializer with a random seed to ensure reproducibility
    weight_initializer = tf.keras.initializers.GlorotNormal(seed=RANDOM_STATE) 
    
    # define input layers
    input_ids_layer = tf.keras.layers.Input(shape=(MAX_LENGTH,), 
                                            name='input_ids', 
                                            dtype='int32')
    input_masks_layer = tf.keras.layers.Input(shape=(MAX_LENGTH,), 
                                                  name='input_attention', 
                                                  dtype='int32')

    

    # tf.tensor representing the hidden-state of the model's last layer
    embedding_layer = transformer([input_ids_layer, input_masks_layer])[0]
    X = Bidirectional(LSTM(MAX_LENGTH, return_sequences=True, dropout=0.2))(embedding_layer)
    X = GlobalMaxPool1D()(X)
    X = Dense(MAX_LENGTH, activation='relu')(X)
    X = Dropout(0.2)(X)
    # Define a single node that makes up the output layer (for binary classification)
    output = Dense(num_classes, 
                                   activation='softmax',
                                   kernel_initializer=weight_initializer,  
                                   kernel_constraint=None,
                                   bias_initializer='zeros'
                                   )(X)
    
    # Define the model
    model = tf.keras.Model([input_ids_layer, input_masks_layer], output)
    
    # Compile the model
    model.compile(tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE), 
                  loss='categorical_crossentropy',
                  metrics=metrics)
    
    return model
```

```python
model = build_model(transformer=distilbert,num_classes=y_train.shape[1])
```

```python
#now we will define a step decay function to reduce our learning rate
def step_decay_schedule(initial_lr=LEARNING_RATE, decay_factor=0.75, step_size=10):
    '''
    Wrapper function to create a LearningRateScheduler with a set step decay schedule
    '''
    def schedule(epoch):
        return initial_lr * (decay_factor ** np.floor(epoch/step_size))
    
    return LearningRateScheduler(schedule)
```

```python
#define callbacks for our model
checkpoint = ModelCheckpoint(filepath = MODEL_PATH + f'BERT/model_{dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}_best_weights.h5', 
                             monitor='val_loss',
                             verbose=1,
                             save_weights_only = True, 
                             save_best_only=True,
                             mode='min')

lr_sched = step_decay_schedule(initial_lr=LEARNING_RATE, decay_factor=0.75, step_size=2)

early_stopping = EarlyStopping(monitor='val_loss', patience=10)
tqdm_callback = tfa.callbacks.TQDMProgressBar()

callbacks = [checkpoint,
             early_stopping,
             lr_sched,
             tqdm_callback]
```

```python
#we will set a large number of epochs as we'll implement early stopping once the model starts overfitting
EPOCHS = 300
BATCH_SIZE = 64
NUM_STEPS = X_train_input_ids.shape[0] // BATCH_SIZE

# Train the model
history = model.fit(
    x = [X_train_input_ids, X_train_input_masks],
    y = y_train,
    epochs = EPOCHS,
    batch_size = BATCH_SIZE,
    steps_per_epoch = NUM_STEPS,
    validation_data = ([X_val_input_ids, X_val_input_masks], y_val),
    verbose=2,
    class_weight=class_weights,
    callbacks=callbacks
)
```

```python
with open(MODEL_PATH + f'BERT/model_{dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}_history.h5', 'wb') as f:
        pickle.dump(history.history, f)
```

```python
# model trains for 1h17m11s and we will now test it on our test data 
# Evaluate the model on the test data using `evaluate`
# print("Evaluate model on test data")
y_pred = model.predict([X_test_input_ids, X_test_input_masks])
```

```python
weighted_auc = metrics.roc_auc_score(y_test, y_pred, 
                            average='weighted',multi_class='ovr')
print(f'Weighted AUC: {weighted_auc}')
```

We can see that our new DistilBERT model is still struggling with the smaller classes some. We can see that classes 2 and 3, which correspond to the Disabled Access and Electrical Inspection Divisions, and are the minority classes still have poor precision, in other words our model is predicting them too often. The challenge here is that there are far more instances of the majority classes, so even a small error rate relative to the majority class still accounts for a large number of false predictions in comparison to the size of the minority class. 

However, recall and precision are a trade off, and optimizing the model for greater precision would result in lower recall which would mean those minority classes would be predicted less. In this case that would mean that people with Disabled Complaints and Electrical Complaints would not have their needs met. 

This is not to say that we couldn't still improve uppon the precision while mainting our recall, possibly by exploring other model architectures. gathering additional data, or exploring data augmentation, but for now I am happy with the model's performances as I have other projects to work on.

We can also see that the validation loss is very volatile, this is likely due to the fact that our validation set is very small compared to the training set and the number of parameters, with one class only having about 100 entries. However, the validation loss is still decreasing and this doesn't effect model performance. Ideally we would have a larger validation set, but because I am already working with a small data set here I didn't want to sacrifice aditional data to be held out in the validation set.
